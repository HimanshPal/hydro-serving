```python
from sklearn import datasets
from sklearn import svm
import numpy as np
import pandas as pd
```

```python
iris = datasets.load_iris()
```

```python
clf = svm.SVC(gamma=0.001, C=100.)
clf.fit(iris.data[:-1], iris.target[:-1])  
```

```python
clf.predict(iris.data[-1:])
```

```python
df = pd.DataFrame(data=np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
input_df = pd.DataFrame(data=iris['data'], columns = iris['feature_names'])
output_df = pd.DataFrame(data=iris['target'], columns = ['target'])
```

```python
pd.concat([input_df, output_df], axis=1, join='inner')
```

```python
iris.data[:2]
```

```python
from sklearn.externals import joblib
import json

joblib.dump(clf, '../models/svm/model.pkl')
incols = list(input_df.columns)
outcols = list(output_df.columns)
metadata = {}
metadata['model'] = type(clf).__name__
metadata['inputs'] = incols
metadata['outputs'] = outcols
with open('../models/svm/metadata.json', 'w') as fp:
    json.dump(metadata, fp)
```
